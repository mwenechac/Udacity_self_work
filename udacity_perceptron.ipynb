## AND Gate

```
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import numpy as np
```

```
#declaring values
and_gate=[(0,0),(0,1),(1,0),(1,1)]
and_gate_output=[0,0,0,1]
```

```
#the space for w1,w2 and b for which we will be looking for values
w1=np.linspace(0,10,20)
w2=np.linspace(0,10,20)
b=np.linspace(-10,0,20)
XX,YY,ZZ=np.meshgrid(w1,w2,b)
YYZZCC=np.zeros(XX.shape)
```

```
def and_perceptron(w1,w2,b):
  result=True
  for q,w in zip(and_gate,and_gate_output):
    result=result*(((q[0]*w1+q[1]*w2+b)>=0)==w)
  return result
```

```
#finding the result for our space
for i in range(w1.size):
  for j in range(w2.size):
    for k in range(b.size):
      YYZZCC[i,j,k]=and_perceptron(w1[i],w2[j],b[k])
```

```
#since we are looking to plot just the "right" values, converting 0's to nan (matplotlib does not plot nans)
def zero_to_nan(values):
    return [float('nan') if x==0 else x for x in values]
res=zero_to_nan(YYZZCC.reshape(1,-1)[0])
```

```
mm=np.where(YYZZCC==1)
```

```
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(figsize=(15,7))
ax = plt.axes(projection='3d')
ax.scatter3D(XX,YY,ZZ,c=res)
plt.xlabel('w1')
plt.ylabel('w2')
ax.set_zlabel('b')
ax.view_init(20,10)
```

```
fig = plt.figure(figsize=(15,7))
ax1 = plt.axes(projection='3d')
ax1.plot_trisurf(w1[mm[0]],w2[mm[1]] ,b[mm[2]],edgecolor='none')
plt.xlabel('w1')
plt.ylabel('w2')
ax1.set_zlabel('b')
ax1.view_init(20,10)
```

## OR perceptron

```
or_gate=[(0,0),(0,1),(1,0),(1,1)]
or_gate_output=[0,1,1,1]
```

```
def or_perceptron(w1,w2,b):
  result=True
  for q,w in zip(or_gate,or_gate_output):
    result=result*(((q[0]*w1+q[1]*w2+b)>=0)==w)
  return result
```

```
for i in range(w1.size):
  for j in range(w2.size):
    for k in range(b.size):
      YYZZCC[i,j,k]=or_perceptron(w1[i],w2[j],b[k])
res=zero_to_nan(YYZZCC.reshape(1,-1)[0])
mm=np.where(YYZZCC==1)
fig = plt.figure(figsize=(15,7))
ax = plt.axes(projection='3d')
ax.scatter3D(XX,YY,ZZ,c=res)
plt.xlabel('w1')
plt.ylabel('w2')
ax.set_zlabel('b')
ax.view_init(20,10)
```

```
fig = plt.figure(figsize=(15,7))
ax1 = plt.axes(projection='3d')
ax1.plot_trisurf(w1[mm[0]],w2[mm[1]] ,b[mm[2]],edgecolor='none')
plt.xlabel('w1')
plt.ylabel('w2')
ax1.set_zlabel('b')
ax1.view_init(20,10)
```
